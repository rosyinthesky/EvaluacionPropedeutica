```python
import math

def menu (): 
    print("Menu:")
    print("1. Suma")
    print("2. Resta")  
    print("3. Multiplicacion")  
    print("4. Division")    
    print("5. Raiz de n") 
    print("6. Exponente de n")
    print("7. Seno") 
    print("8. Coseno")
    print("9. Tangente")
    print("10. Salir")
    return

def suma(num1, num2):
    return num1 + num2
def resta(num1,num2):
    return num1-num2
def multi(num1, num2):
    return num1*num2
def div(num1, num2):
    return num1/num2
def raiz (num):
    return math.sqrt(num)
def exponte(num1,num2):
    return math.pow(num1,num2)
def seno(num):
    return math.sin(num)
def coseno(num):
    return math.cos(num)
def tan(num):
    return math.tan(num)

def calculadora():
    menu()
    while True:
        opc = int (input("Ingrese su opción: "))
        if opc == 1:
                num1 = int(input("Ingresa el primer número: "))
                num2 = int(input("Ingresa el segundo número: "))
                print("La suma de ambos números es: ", suma(num1,num2))
                continue
        elif opc == 2:
            num1 = int(input("Ingresa el primer número: "))
            num2 = int(input("Ingresa el segundo número: "))
            print("La resta de ambos números es: ", resta(num1, num2))
            continue
        elif opc == 3:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print("La multiplicación de ambos números es: ", multi(num1, num2))
            continue
        elif opc == 4:
            num1 = float(input("Ingresa el primer número: "))
            num2 = float(input("Ingresa el segundo número: "))
            print("La división de ambos números es: ", div(num1, num2))
            continue
        elif opc == 5:
            num = int(input("Ingresa el número que desee obtener su raíz"))
            print("La raíz de", num , "es: ", raiz(num))
            continue
        elif opc == 6:
            num1 = float (input("Ingrese el número base: "))
            num2 = float (input("Ingrese el exponencial: "))
            print("El número ", num1, " a la potencia ", num2, "es: ", exponte(num1, num2))
            continue
        elif opc == 7:
            num = float(input("Ingrese el número al que desea sacar el seno en radianes: "))
            print("El seno en radianes es: ", seno(num))
            continue
        elif opc == 8:
            num = float(input("Ingrese el número al que desea sacar el coseseno en radianes: "))
            print("El coseno en radianes es: ", coseno(num))
            continue
        elif opc == 9:
            num = float(input("Ingrese el número al que desea sacar la tangente en radianes: "))
            print("La tangente en radianes es: ", tan(num))
            continue
        elif opc == 10:
            break
    return

calculadora()
```
